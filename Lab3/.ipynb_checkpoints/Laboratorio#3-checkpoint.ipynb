1. ¿Cuál es el tamaño en bytes de un array de 1000 elementos de tipo booleano?

```python
import numpy as np 

bool_array = np.ones(1000).astype(bool) 

# Un elemento tipo booleano ocupa un byte según https://numpy.org/doc/stable/reference/arrays.scalars.html#numpy.bool_

Byte_size = len(bool_array)

print(f'Un array de 1000 elementos booleanos tiene un tamaño de {Byte_size} bytes')
```

5. Tengo valores de coordenadas (x, y) en las columnas del array

   1.33 , 4.5
   
   
   30.0 , 10.7
   
   
   70.2 , 0.5
   
   Agregue a este array las coordenadas (37.1, -3.6).
   Muestre en pantalla las dimensiones del nuevo
  

```python
coordenadas = np.array([[1.33, 4.5],
                        [30.0, 10.7],
                        [70.2, 0.5]])
coordenadas_adicionales = np.array([[37.1,-3.6]])
```

```python
def agregar_coordenadas(arreglo_original, arreglo_adicion, axis=0):
    arreglo_actualizado = np.concatenate((arreglo_original, arreglo_adicion), axis=axis)
    return arreglo_actualizado
```

```python
print(f'Dimensiones del arreglo previo a la adicion {coordenadas.shape}\n')
coordenadas = agregar_coordenadas(coordenadas, coordenadas_adicionales)
print(f'Dimensiones del arreglo posterior a la adicion {coordenadas.shape}\n')
```

 5. Obtenga la media, mediana, desviación estándar, valor máximo y valor mínimo de la población de los municipios de la provincia de Granada.

```python
mun_prov_Granada = nucleospoblacion[nucleospoblacion['Provincia']=='Granada']

media = mun_prov_Granada['Poblacion'].mean()
std =mun_prov_Granada['Poblacion'].std()
max_value =mun_prov_Granada['Poblacion'].max()
min_value =mun_prov_Granada['Poblacion'].min()

print(f'La media de la poblacion en los municipios de la provincia de Granada es: {media}')
print(f'La desviacion estandar de la poblacion en los municipios de la provincia de Granada es: {std}')
print(f'El valor máximo de la poblacion en los municipios de la provincia de Granada es: {max_value}')
print(f'El valor mínimo de la poblacion en los municipios de la provincia de Granada es: {min_value}')

```

6. Realice un histograma con la población de los Municipios para cada una de las provincias.

```python
Provincias = nucleospoblacion['Provincia'].unique()
Provincias

for prov in Provincias:
    Poblacion_provincia = nucleospoblacion[nucleospoblacion['Provincia']== prov]
    plt.hist(Poblacion_provincia['Poblacion'])
    plt.ylabel('Cantidad de Municipios')
    plt.xlabel('Poblacion')
    plt.title(f'Histograma de la Poblacion de los Municipios de la Provincia de {prov}')
    plt.plot()
    plt.show()
```

7. Seleccione al azar cincuenta municipios diferentes de entre los diponibles en el archivo. Asegúrese de que no se repitan. ¿Luego calcule el promedio de la población y la desviación estándar de esto 50 municipios?.

```python
selector = np.zeros_like(nucleospoblacion['Municipio'])

change_pos = np.random.choice(len(selector),50, replace=False) # genera un arreglo de 50 numeros sin repetición entre 0 y el tamaño de filas del dataset 

selector[change_pos] = 1

selector = [bool(value) for value in selector]

municipios_aleatoreos = nucleospoblacion[selector]

mean_muni_al= municipios_aleatoreos['Poblacion'].mean()
std_muni_al =  municipios_aleatoreos['Poblacion'].std()

print(f'El promedio de la población de los 50 municipios selecionados aleatoreamente es: {mean_muni_al}\n')

print(f'La desviación estandar de la población de los 50 municipios selecionados aleatoreamente es: {std_muni_al}')
```


8. Determine los nombres de los Municipios más cercano y más lejano a Madrid. Para ello debe calcular la distancia en todos ellos y Madrid. No considere a Madrid en el análisis ya que la distancia sería cero.
  

```python

x_madrid= float(nucleospoblacion['X'][nucleospoblacion['Municipio']=='Madrid'])
y_madrid= float(nucleospoblacion['Y'][nucleospoblacion['Municipio']=='Madrid'])

nucleospoblacion['Distancia_a_Madrid']=  np.sqrt(((nucleospoblacion['X']-x_madrid)**2) + (nucleospoblacion['Y']-y_madrid)**2)

nucleospoblacion_sin_madrid = nucleospoblacion[nucleospoblacion['Municipio'] != 'Madrid']
nucleospoblacion_sin_madrid
```

```python
Municipio_mas_lejano = nucleospoblacion_sin_madrid[nucleospoblacion_sin_madrid['Distancia_a_Madrid'] 
                                                    ==nucleospoblacion_sin_madrid['Distancia_a_Madrid'].max()]
```

```python
print(f'El municipio mas lejano a Madrid es {Municipio_mas_lejano["Municipio"].unique()[0]}a  una distancia de {float(Municipio_mas_lejano["Distancia_a_Madrid"])}')
```

```python
Municipio_mas_cercano = nucleospoblacion_sin_madrid[nucleospoblacion_sin_madrid['Distancia_a_Madrid'] 
                                                    ==nucleospoblacion_sin_madrid['Distancia_a_Madrid'].min()]
```

```python
print(f'El municipio mas cercano a Madrid es {Municipio_mas_cercano["Municipio"].unique()[0]}a  una distancia de {float(Municipio_mas_cercano["Distancia_a_Madrid"])}')
```
